```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
SMALL_SIZE  = 12
MEDIUM_SIZE = 15
BIGGER_SIZE = 20

plt.rc('font', size        = BIGGER_SIZE)   # controls default text sizes
plt.rc('axes', titlesize   = BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize   = BIGGER_SIZE)   # fontsize of the x and y labels
plt.rc('xtick', labelsize  = MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize  = MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize  = SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize = BIGGER_SIZE)   # fontsize of the figure title
```

# MNIST

```python
mnist_iid = pd.read_csv("mnist_tests/mnist_iid.txt", header = None)
mnist_iid.columns = ["round", "accuracy", "loss"]

mnist_uniform = pd.read_csv("mnist_tests/mnist_uniform.txt", header = None)
mnist_uniform.columns = ["round", "accuracy", "loss"]

mnist_90percent = pd.read_csv("mnist_tests/mnist_90percent.txt", header = None)
mnist_90percent.columns = ["round", "accuracy", "loss"]

mnist_100percent = pd.read_csv("mnist_tests/mnist_100percent_10steps.txt", header = None)
mnist_100percent.columns = ["round", "accuracy", "loss"]
```

```python
mnist_accuracies = pd.concat([mnist_iid["accuracy"], mnist_uniform["accuracy"], 
                        mnist_90percent["accuracy"], mnist_100percent["accuracy"]], axis = 1)

mnist_accuracies.columns = ["Non-federated", "10% from each class", "90% from each class", "100% from each class"]
```

```python
mnist_losses = pd.concat([mnist_iid["loss"], mnist_uniform["loss"], 
                        mnist_90percent["loss"], mnist_100percent["loss"]], axis = 1)

mnist_losses.columns = ["Non-federated", "10% from each class", "90% from each class", "100% from each class"]
```

```python
ax = mnist_accuracies.plot(figsize = (18, 5))

ax.set_xlabel("Round", fontsize = 20)
ax.set_ylabel("Accuracy", fontsize = 20)

plt.tight_layout()
```

```python
ax = mnist_losses.plot(figsize = (18, 5))

ax.set_xlabel("Round", fontsize = 20)
ax.set_ylabel("Loss", fontsize = 20)

plt.tight_layout()
```
